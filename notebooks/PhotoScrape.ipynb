```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://photos.google.com/share/AF1QipNssJQozIaEYt7WeG00S0T9FOB-QYyhxLsD9_NZGFqgLkMESSKEpx-cWE1MySbo1Q?key=bXNGbHYyM0hfb3BKMFFrODFkbXc0SVlxanlUbV9n')
```

```python
driver.execute_script("document.querySelector('.B6Rt6d.zcLWac.eejsDc').scrollTo(0,0)") 

all_links = set([])
scroll_top = -1
new_scroll_top = 0
counter = 0
while scroll_top != new_scroll_top and counter < 200:
    counter += 1
    content = driver.find_elements_by_class_name('xfzUCb')
    a = content[0]
    b = a.find_elements_by_tag_name('a')
    links = [c.get_attribute('href') for c in b]
    scrollable_thing = driver.find_element_by_css_selector('.B6Rt6d.zcLWac.eejsDc')
    driver.execute_script("document.querySelector('.B6Rt6d.zcLWac.eejsDc').scrollBy(0,500)")
    scroll_top = new_scroll_top
    new_scroll_top = scrollable_thing.get_attribute('scrollTop')
    all_links.update(links)
    
len(all_links)
```

```python
urls = []
count = 0
for link in all_links:
    count += 1
    print("{0} of {1}".format(count, len(all_links)))
    driver.get(link)
    content = driver.find_element_by_class_name('SzDcob')
    url = content.get_attribute('src')
    urls.append(url)
```
